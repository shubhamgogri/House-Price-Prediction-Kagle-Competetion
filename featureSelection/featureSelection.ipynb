```python
import pandas as pd
import numpy as np

pd.pandas.set_option('display.max_columns', None)
```

```python
url = 'https://raw.githubusercontent.com/shubhamgogri/House-Price-Prediction-Kagle-Competetion/main/FeatureEng/featured_train.csv'
X_train = pd.read_csv(url, index_col=0)
X_train  = X_train.reset_index()
```

```python
X_train.head()
```

```python
X_train.drop(['Id'], inplace = True, axis =1)
```

```python
y_test = X_train.iloc[:, :1]
y_test.head()
```

```python
X_train.drop('SalePrice', inplace = True, axis = 1)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

lasso = Lasso()
```

```python
feature_sel = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel.fit(X_train, y_test)
```

```python
feature_sel.get_support()
```

```python
selected = X_train.columns[feature_sel.get_support()]
```

```python
print('Total features', len(X_train.columns))
```

```python
X_train = X_train[selected]
print('Selected Features', len(selected) )
```
