**Feature Engineering**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.pandas.set_option('display.max_columns', None)
```

```python
train = pd.read_html('https://github.com/shubhamgogri/House-Price-Prediction-Kagle-Competetion/blob/main/train.csv')
train = train[0]
train.drop('Unnamed: 0', axis = 1, inplace = True)
train.head()
```

```python
train.shape
```

```python
features_nan = [feature for feature in train.columns if train[feature].isnull().sum() >1 and train[feature].dtypes == 'O']

for feature in features_nan:
    print(feature, np.round(train[feature].isnull().mean(), 4))
```

Replace the missing values with a new label

Categorical features are handled by adding Missing

```python
def replace_cat_features(train, features_nan):
  data = train.copy()
  data[features_nan] = data[features_nan].fillna('Missing')
  return data

train = replace_cat_features(train, features_nan)
train[features_nan].isnull().sum()
```

```python
train.head()
```

For Numerical Variables

```python
num_feature = [feature for feature in train.columns if train[feature].isnull().sum() >1 and train[feature].dtypes != 'O']

for feature in num_feature:
    print(feature, np.round(train[feature].isnull().mean(), 4))
```

When there are lot of outliers replace it with median or mode 

```python
for feature in num_feature:
  median_value = train[feature].median()
  # creating a new column for the feature with nan indicating the 
  # 1 for nan and  0 for not having nan value
  train[feature+ 'nan'] = np.where(train[feature].isnull(), 1 , 0)
  # replace the nan with median
  train[feature].fillna(median_value, inplace =True)

train[num_feature].isnull().sum()  
```

```python
train.head()
```

Temporal Variables

they are handled by counting the 

difference between the year feature and 

yrSold feature

```python
for feature in ['YearBuilt', 'YearRemodAdd',  'GarageYrBlt']:
  train[feature] = train['YrSold'] - train[feature]

train[['YearBuilt', 'YearRemodAdd',  'GarageYrBlt']].head()
```

Only For Continuous variables

skewed = not a guassian distribution

Many numerical Features had skewed values
in which only the features with value other than zero are handled by applying log function.

Discrete - unique values are less than 25

continuous - unique values are more than 25

Here out of 16 continuous features only 5 are treated 

because all other features have value 0

**LOG NORMAL DISTRIBUTION**

```python
num_feature = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_feature:
  train[feature] = np.log(train[feature])
```

```python
train.head()
# hence the log values are converted from normal values
```

**RARE CATEGORICAL FEATURES**

```python
categorical_feat = [feature for feature in train.columns if( train[feature].dtypes == 'O')]
categorical_feat
```

```python
for feat in categorical_feat:
  print(feat  , len(train[feat].unique()))
```

```python
for feat in categorical_feat:
  temp = train.groupby(feat)['SalePrice'].count()/ len(train)
  temp_df = temp[temp>0.01].index
  train[feat] = np.where(train[feat].isin(temp_df), train[feat], 'Rare_var')
```

```python
train['MSZoning'].unique()
```

```python
train.head()
```

```python
# need to map the categorical_features values 
# cant use dummies as it the columns are more
# just map using enumerate function
# doesn't have the sorted values 
for feature in categorical_feat:
    labels_ordered = (train[feature].unique())
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    train[feature] = train[feature].map(labels_ordered)
    # print(labels_ordered)
```

```python
train['BsmtFinType1'].unique()
```

```python
for feature in categorical_feat:
  # gets all the unique values that are sorted 
    labels_ordered= train.groupby([feature])['SalePrice'].mean().sort_values().index
    # print('labels_ordered', labels_ordered)
    # provide key to each of the values in list
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    print('labels_ordered', labels_ordered)
    # update the features
    # train[feature]=train[feature].map(labels_ordered)
```

```python
train.head()
```

**Feature SCaling**

```python
feature_scale = [feature for feature in train.columns if feature not in ['Id', 'SalePrice']]
len(feature_scale)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(train[feature_scale])
```

```python
scaler.transform(train[feature_scale])
```

```python
train.head()
```

Create a new dataframe using the scaled features 

```python
train = pd.concat([train[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(train[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
train.head()
```
